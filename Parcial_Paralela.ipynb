# ***Contexto***
Python es un lenguaje de programaciòn interpretado de alto nivel que le permite al programador flexibilidad, comodidad y escritura dinàmica a la hora de desarrolar sus còdigos fuente. Sin embargo al ser un lenguaje interpretado està optimizado para hacer la vida màs fàcil al programador, lo que significa una carga adicional a la màquina ya que necesita un interprete para traducir su còdigo.En cambio, el lenguaje compilado como C està optimizado para el momento de la ejecuciòn, lo que quiere decir que  es mucho màs ràpido en comparaciòn al interpretado, ya que al ejecutar el proceso de traducciòn del còdigo a còdigo de bajo nivel o còdigo màquina aumenta la sobrecarga y puede ocasionar que el programa sea màs lento.


#***Resumen***


La presente pràctica de laboratorio se hace con el fin de desarrollar una bateria de experimentos significativa para automatizar y optimizar el tiempo de ejecuciòn de 4 programas que son codificados en lenguaje Python. Se agrega la extensiòn Cython, se ejecutan los programas 30 veces cada uno y se almacenan los tiempos obtenidos por programa en archivos csv.



Como se menciona previamente,para el desarrollo de este laboratorio se agrega la extenciòn del lenguaje C a Python denominado Cython, la cual permite usar librerias y variables de tipo C, combinando la velocidad de C con la flexibilidad de sintaxis de Python.



Al momento de usar Cython correctamente se crearon 6 archivos principales por cada programa:

**1.** Archivo **cy_nombre.pyx**, contiene el còdigo en Cython.

**2.** Archivo **py_nombre.py**, contiene el còdigo en Python.

**3.** Archivo **setup.py** encargado de generar la extensiòn de Cython al compilarlo.

**4.** Archivo **performance.py**, encargado de llamar al fichero generado por el setup.py para poder importar el còdigo de Cython como si fuera una librerìa y asi poder acceder a sus funciones.

**5.** Archivo **Makefile**, encargado de automatizar la creaciòn y depuraciòn de los archivos ejecutables.

**6.** Archivo **prueba_nombre.py**, encargado de almacenar los tiempos de ejecuciòn obtenidos en ambos lenguajes en archivos csv.

Posterior a esto se grafican los resultados obtenidos por cada programa, obteniendo que en la mayoria de los casos Cython presenta un rendimiento mucho mejor al de Python, resaltando que en 2 de los programas implementados Cython tuvo una mejoria en promedio del 60% y que Cython presenta un ruido gaussiano mucho menor al de Python en la ejecuciòn de los programas.
